Removing OUTLIERS

```python
lower=0
upper=23000
df=df.loc[(df['Amount']>=lower)&(df['Amount']<= upper),:]
plt.figure(figsize=(15,5))
sns.boxplot(x='Amount',data=df)
plt.title('Box Plot of Amount (After Outliers Removed)')
```

```python
number_of_outliers=df[(df['Amount']<lower) | (df['Amount']>upper)].shape[0]
print("Number of outliers:", number_of_outliers)
```

ENCODING

```python
df1=pd.get_dummies(df[['Gender','Age Group','State','Zone','Occupation','Product_Category']],drop_first=True)
df1
```

```python
df_n=pd.concat([df,df1],axis=1)
```

```python
df_n=df_n.drop(['Gender', 'Age Group','State','Zone','Occupation','Product_Category'],axis=1)
```

SPLITTING X AND Y

```python
x=df_n.drop('Amount',axis=1)
x
```

```python
y=df_n['Amount']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

MODEL CREATION

```python
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(x_train,y_train)
```

```python
y_pred=regr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error,mean_absolute_error
print("R2_score is : ",r2_score(y_test,y_pred))
print("MAPE is : ",mean_absolute_percentage_error(y_test,y_pred))
```
